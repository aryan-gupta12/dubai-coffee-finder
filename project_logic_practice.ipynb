```python
from geopy.distance import geodesic
user_location = (25.175,55.275)
spot_location = (25.152942,55.295840)

round(geodesic(user_location,spot_location).km,2)
```

```python
def calculate_distance(user_location,row):
    lat = row["lat"]
    lng = row["lng"]
    spot_location = (lat,lng)
    dis = geodesic(user_location,spot_location).km
    return round(dis,2)

def get_row(row):
    return calculate_distance(user_location,row)


df["distance_km"] = df.apply(get_row,axis=1)
```

```python
from datetime import datetime

datetime.now().strftime("%H:%M")
```

```python
def is_open(row):
   current_time =  datetime.now().strftime("%H:%M")
   open_time = row["opening_time"]
   close_time = row["closing_time"]

   if open_time <= current_time <= close_time:
      return "Open"
   else:
      return "Closed"

   
```
